# Direct Preference Optimization (DPO)

This notebook demonstrates the Direct Preference Optimization (DPO) stage of the Customer Support Assistant.

DPO aligns the Supervised Fine-Tuned (SFT) model with preferred responses using a preference dataset containing **chosen** and **rejected** answers.

Unlike Supervised Fine-Tuning, DPO directly optimizes the model to generate responses that better match human preferences.

# Training Configuration Explanation

The DPO stage was executed using the **Google Colab Free Tier (T4 GPU)**.

Since DPO is computationally more expensive than SFT, a smaller subset of the preference dataset was used.

| Parameter | Value | Purpose |
|-----------|------:|---------|
| Batch Size | 2 | Number of preference pairs processed in each training step |
| Training Subset Size | 150 | Number of preference pairs used during DPO |
| Fine-Tuning Method | DPO | Preference optimization |
| Starting Model | SFT Adapter | Continues training from the SFT checkpoint |

Using a smaller subset reduces training time while demonstrating the complete Direct Preference Optimization workflow.

# SFT vs DPO

| Supervised Fine-Tuning (SFT) | Direct Preference Optimization (DPO) |
|------------------------------|---------------------------------------|
| Learns from instruction-response pairs | Learns from preferred and rejected responses |
| Optimizes next-token prediction | Optimizes preference alignment |
| Produces an instruction-following model | Produces a preference-aligned model |
| First stage of alignment | Final alignment stage |

Together, SFT and DPO provide a complete post-training pipeline for adapting the base language model to the Customer Support domain.
