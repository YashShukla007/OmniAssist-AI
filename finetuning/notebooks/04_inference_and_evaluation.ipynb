# Evaluation Metrics

The following metrics are used during evaluation.

| Metric | Purpose |
|---------|---------|
| BLEU | Measures n-gram overlap |
| ROUGE-1 | Measures unigram overlap |
| ROUGE-2 | Measures bigram overlap |
| ROUGE-L | Measures longest common subsequence |
| BERTScore | Measures semantic similarity |

Together these metrics provide a balanced evaluation of lexical similarity and semantic correctness.
